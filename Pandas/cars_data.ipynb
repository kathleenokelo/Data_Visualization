```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import squarify
```

```python
#1.Import the dataset and assign it to variable named cars
file_path = r'C:\Users\Kathleen\Downloads\car_prices.csv'
```

```python
cars = pd.read_csv(file_path)
cars
```

```python
#2.Identify whether there are any missing values in the data then drop them

cars.isnull().value_counts()
```

```python
cars_dropped = cars.dropna()
cars_dropped
```

```python
#4.What are the top 5 most expensive cars

cars.nlargest(5,'sellingprice')
```

```python
#5.What are the most common car makes and models

cars_make = cars['make'].value_counts()
cars_model = cars['model'].value_counts()

print('The most common car makes are\n',cars_make)
print('The most common car model \n',cars_model)
```

```python
cars.head()
```

```python
#6.Which car models have the highest and lowest average prices
avg_price= cars.groupby(by='model')['sellingprice'].mean()

max_avg_model = avg_price.max()
min_avg_model = avg_price.min()


max_avg = cars.groupby(by='model')['sellingprice'].mean().idxmax()
min_avg = cars.groupby(by='model')['sellingprice'].mean().idxmin()

print(f'{max_avg} has the highest average price of {max_avg_model} dollars')
print(f'{min_avg} has the lowest average price of {min_avg_model} dollars')
```

```python
#7.How does car prices vary with different attributes

y1 = cars.groupby('make')['sellingprice'].mean().iloc[:10,]
y2 = cars.groupby('year')['sellingprice'].mean()
y3= cars.groupby('model')['sellingprice'].mean().iloc[:10,]
y4 = cars.groupby('odometer')['sellingprice'].mean().iloc[:10,]

x = cars['sellingprice']


fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(30,15))

#Plot by make
ax[0,0].bar(y1.index,y1.values)
ax[0,0].set_title('Average prices by make')
ax[0,0].set_ylabel('make')
ax[0,0].set_xlabel('Average Selling Price')
#plt.tick_params(axis='x',labelrotation=45)

#Plot by year
ax[0,1].plot(y2.index,y2.values,color='Black')
ax[0,1].set_title('Average prices by year')
ax[0,1].set_xlabel('Year')
ax[0,1].set_ylabel('Average Selling Price')

#Plot by Model
ax[1,0].bar(y3.index,y3.values)
ax[1,0].set_title('Average prices by model')
ax[1,0].set_xlabel('Model')
ax[1,0].set_ylabel('Average Selling Price')

#Plot by MMR
ax[1,1].bar(y3.index,y3.values)
ax[1,1].set_title('Average prices by mileage')
ax[1,1].set_xlabel('Mileage')
ax[1,1].set_ylabel('Average Selling Price')

plt.tight_layout()


plt.show()


```

```python
cars['saledate'] = pd.to_datetime(cars['saledate'],errors='coerce',utc=True)

cars = cars.dropna(subset='saledate')

cars.dtypes
```

```python
#8.Is there any seasonal trend in car sales prices

import datetime as dt

cars['salemonth'] = pd.to_datetime(cars['saledate']).dt.month
mean = cars.groupby('salemonth')['sellingprice'].mean()
mean
```

```python
plt.figure(figsize=(15,6))
plt.plot(mean.index,mean.values)
plt.xlabel('MONTHS')
plt.ylabel('SELLING PRICE')
plt.title('Trend of prices by months')
plt.show()
```

```python
#9.What is the relationship between car price and mileage

plt.figure(figsize=(15,6))
plt.scatter(cars['odometer'],cars['sellingprice'],alpha=0.5)
plt.title('Relationship between Car prices and Mileage')
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.grid(True)
plt.show
```
